```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
archivo = "train.csv"
df=pd.read_csv(archivo)
```

```python
df.head()       # Muestra las primeras filas
```

```python
df.info()       # Información general del dataframe
```

```python
df.describe()   # Estadísticas básicas
```

```python
df.isnull().sum() #Verificar los valores nulos
```

```python
# Distribución por programa académico
df['ESTU_PRGM_ACADEMICO'].value_counts().head(10).plot(kind='bar', figsize=(10, 5))
```

```python
# Estudiantes por departamento
df['ESTU_PRGM_DEPARTAMENTO'].value_counts().plot(kind='barh', figsize=(10, 6))
```

```python
# Cantidad de estudiantes por rendimiento
df['RENDIMIENTO_GLOBAL'].value_counts().plot.pie(autopct='%1.1f%%', figsize=(6, 6))
```

```python
# Comparacion de estudiantes con acceso a internet y estudiantes que poseen coputador
acceso = df.groupby(['FAMI_TIENEINTERNET', 'FAMI_TIENECOMPUTADOR']).size().unstack()

acceso.plot(kind='bar', stacked=True, figsize=(8,6), colormap='Set2')
plt.title('Acceso a Internet y Computador')
plt.xlabel('¿Tiene Internet?')
plt.ylabel('Número de estudiantes')
plt.xticks(rotation=0)
plt.legend(title='¿Tiene Computador?')
plt.tight_layout()
plt.show()
```

```python
#Comparacion del rendimiento segun el nivel educativo de la madre
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='FAMI_EDUCACIONMADRE', hue='RENDIMIENTO_GLOBAL', palette='Set3')
plt.title('Rendimiento según educación de la madre')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Educación Madre')
plt.ylabel('Cantidad')
plt.tight_layout()
plt.show()
```
